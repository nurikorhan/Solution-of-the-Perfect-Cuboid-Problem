```python
my_diag_1_list=[]
from math import sqrt
my_base=3#the first number will be 3**x. it could also be 5,7,11,13 etc.
for my_power in range(3,15):
    my_side=my_base**my_power
    my_second_side_list=[]
    my_diag_1_list=[]
    for my_coefficient_power in range (my_power):
        my_side_power=my_power-my_coefficient_power
        my_coefficient=my_base**my_coefficient_power
        my_second_side=my_coefficient*((my_side/my_coefficient)**2-1)/2
        my_third_side=my_coefficient*((my_side/my_coefficient)**2+1)/2
        my_second_side_list.append(my_second_side)
        my_diag_1_list.append(my_third_side)
    for i,my_second_side_alt in enumerate(my_second_side_list):
        for j,my_diag_1_alt in enumerate(my_diag_1_list):
            if i!=j:
                my_object_diagonal=sqrt((my_second_side_alt)**2+(my_diag_1_alt)**2)
                my_second_side_org=my_second_side_list[j]
                my_diag_2=sqrt((my_second_side_alt)**2+(my_second_side_org)**2)
                if (my_object_diagonal==int(my_object_diagonal)):
                    if (my_diag_2==int(my_diag_2)):
                        print("my_power=",my_power)
                        print(my_side,my_second_side_alt,my_second_side_org)
                        print("my_diag_1=",my_diag_1_alt)
                        print("my_diag_2=",my_diag_2)
                        print("my_object_diagonal",my_object_diagonal)
                        
```
